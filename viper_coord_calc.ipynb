```python
L1,L2,x,y,st1,st2,ct1,ct2 = var('L1,L2,x,y,st1,st2,ct1,ct2')
eq = [x == L1*ct1 + L2*ct2, 
      y == L1*st1 + L2*st2,
     st1*st1 + ct1*ct1 == 1,
     st2*st2 + ct2*ct2 == 1]
sol = solve(eq, [st1,st2,ct1,ct2],solution_dict=True)[0]
```

```python
show(sol)
```

```python
vals = {L1: 570.2830296,
    L2:  629.863324,
    x: 0,
    y: 500,
      }
theta1 = arctan2(sol[ct1], sol[st1])
theta2 = arctan2(sol[ct2], sol[st2]) - theta1
```

```python
for pt in [(500,0), (-500,0), ( 500, 1000), (0, 500), (0, 500.1), (0, 505.0),(-500,1000),(0,350),(0,450)]:
    vals = {L1: 570.2830296,
        L2:  629.863324,
        x: pt[0],
        y: pt[1],
          }    
    print(theta1.subs(vals).n(), theta2.subs(vals).n())
```
