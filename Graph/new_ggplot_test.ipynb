## sageカーネルでは表示に不具合がある
ggplotがjupyterに対応したようなので、確認してみました。

残念ながら、sageカーネルでは表示がダブって表示される以前の不具合が残っていました。
そこで、load_extでsageも使えるようにした状態で、ggplotの例題を実行してみます。

```python
%load_ext sage
```

```python
%matplotlib inline
from ggplot import *
```

```python
ggplot(mpg, aes(x='cty', y='hwy')) + \
    geom_point() + \
    ggtitle("City vs. Highway Miles per Gallon")
```

```python
ggplot(meat, aes(x='date', y='beef')) + \
    geom_point() + \
    stat_smooth(ma=12, color='coral')
```

```python
ggplot(diamonds, aes(x='price', fill='clarity')) + geom_histogram()
```

```python
ggplot(meat, aes(x='date', y='pd.rolling_mean(beef, 12)')) + \
    geom_line()
```
